```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix


%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
raw_data_count = recent_grads.shape[0]

# Dropping rows with null values
recent_grads = recent_grads.dropna(axis=0)
clean_data_count = recent_grads.shape[0]

recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(5,10))
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs. Sample_size', figsize=(5,10))
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full Time vs. Median', figsize=(5,10))
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment rate', figsize=(5,10))
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(5,10))
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(5,10))

```

```python
major_categories = recent_grads['Major_category'].unique()
columns = ['Men', 'Women']

df = pd.DataFrame(index=major_categories, columns=columns)

for category in major_categories:
    a = recent_grads[recent_grads['Major_category'] == category]
    rows = a.sum()
    df.loc[category] = rows

ax = df.plot.bar(stacked=True)
ax.set_title('Men vs Women in each category')
ax.set_xticklabels(major_categories, rotation=90)
#rint(df.head())
```
